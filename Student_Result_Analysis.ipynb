# gender distribution

```python
plt.figure(figsize=(4,3))
ax=sns.countplot(data=df,x='Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
##from the above chart female is more than number of males
```

```python
gb=df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
```

```python
print(gb)
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(gb,annot=True)
plt.show
```

```python
# from the about chart we have concluded the education of parents have good impact
```
